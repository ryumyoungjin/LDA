```python
bamin = pd.read_csv( '배민review.txt', sep = "\t",encoding="cp949")
cupang = pd.read_csv( '쿠팡이츠review.txt', sep = "\t",encoding="cp949")
```

```python
bamin.head()
```

```python
bamin_s = bamin.sort_values(by=["ratings"], ascending=False)
bamin_s
```

```python
bamin_s['comment_len'] = bamin_s['comment'].apply(lambda x : len(x))
bamin_s_5up = bamin_s[bamin_s['comment_len'] > 5]
bamin_s_5up
```

```python
bamin_s_5up.to_csv('bamin_s_5up.txt',index=False,sep='\t')
```
